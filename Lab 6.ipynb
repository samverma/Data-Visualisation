# Importing the dataset and displaying the top 5 rows

```python
import seaborn as sns
df = sns.load_dataset('flights')
```

```python
df.head()
```

# Pivoting the data which means to reshape the data based on column values.

```python
dfpivot = df.pivot("month","year","passengers")
```

```python
a , ax =plt.subplots(figsize=(9, 6))
sns.heatmap(dfpivot, annot =True, fmt="d", linewidths=.5, ax=ax)
plt.show()
```

```python
sns.set(style="darkgrid")
ax = sns.barplot(x="month",y="passengers",data=df,ci=None)
plt.show()
```

Through this barplot, I understand that the number of passengers taking flights in the month if July and August are the highest. Likewise we can see the trend for all the months. The least passengers were in the month of February and November.

```python
df= df.sort_values(by="year", ascending =False)
sns.set(style="darkgrid")
ax = sns.barplot(x="year",y="passengers",data=df,ci=None)
plt.show()
```
